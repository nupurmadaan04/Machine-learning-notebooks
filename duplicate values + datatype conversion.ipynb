```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

HANDLE DUPLICATES

```python
# check for duplicates
df['dup']=df['Sex'].duplicated()
df['dup_name']=df['Name'].duplicated()
```

```python
df
```

```python
# drop duplicates
# drop duplicate rows with same name
df.drop_duplicates(subset='Name',inplace=True)
# drop duplicate with same Sex
df.drop_duplicates(subset='Sex',inplace=True)
```

```python
df
```

CONVERT DATA TYPES

```python
# convert data type
df['FARE']=df['Fare'].astype(int)
```

```python
df
```

```python
# error free conversion, errors = 'coerce'
# fill error value with NaN value
df['Age'] = pd.to_numeric('float', errors = 'coerce')
```

```python
# convert_dtypes automatic conversion, intelligent for whole overall cleanup, not for some selected columns
df = df.convert_dtypes()
```
