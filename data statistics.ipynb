```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.head(10)
```

```python
df.describe()
```

```python
min_r = df['Fare'].min()
print(min_r)
```

```python
max_r = df['Fare'].max()
print(max_r)
```

```python
range = max_r - min_r
print("range = ", {range})
```

```python
Fare_mean = df['Fare'].mean()
print("mean = ", {Fare_mean})
```

```python
# mad is mean absolute deviation - deviation of each point from mean
# summ((xi - xmean)/n)
mad = np.mean(np.abs(df['Fare']- Fare_mean))
print(mad)
```

```python
# sd is standard deviation , means used to find deviation when spread is normally distributed
# ddof =0 or default , means population data
# ddof = 1 , means sample data (ddof = delta degree of freedom)
sd = np.std(df['Fare'])
print(sd)

sd1 = np.std(df['Fare'],ddof = 1)
print(sd1)
```

```python
# var , variance used to data kitna spread h mean se but sqaured form
# var = (std)^2
vr = np.var(df['Fare'])
print(vr)

vr1 = np.varvr = np.var(df['Fare'],ddof = 1)
print(vr1)
```
