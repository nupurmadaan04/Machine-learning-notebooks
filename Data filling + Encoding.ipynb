```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Fare'].fillna(df['Fare'].median(), inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
```

```python
# do backward filling , lower ones to upper
# ffill oppostie of this
df['Cabin'].fillna(method='bfill')

# axis = 1 , for columns df['Cabin'].fillna(method='bfill',axis=1)
```

```python
from sklearn.impute import SimpleImputer

# Mean → for numerical data (no outliers ideally)
imputer_mean = SimpleImputer(strategy='mean')
df['Age'] = imputer_mean.fit_transform(df[['Age']])

# Median → for numerical data (best when outliers present)
imputer_median = SimpleImputer(strategy='median')
df['Age'] = imputer_median.fit_transform(df[['Age']])

# Mode (most frequent) → for categorical data
imputer_mode = SimpleImputer(strategy='most_frequent')
df['Embarked'] = imputer_mode.fit_transform(df[['Embarked']])
```

```python
df.isnull().sum()
```

```python
# one hot encoder , less columns using pandas else we can do by sklearn also
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)

# sklearn
#from sklearn.preprocessing import OneHotEncoder
#ohe = OneHotEncoder(sparse=False, drop='first')
#encoded = ohe.fit_transform(df[['Sex']])
# convert to DataFrame
#encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(['Sex']))
# merge back
#df = pd.concat([df, encoded_df], axis=1)
#df.drop('Sex', axis=1, inplace=True)
```

```python
# used when categories have no order like A,B,C no specific order
# Label Encoder gives random values like 4,2,1,5 etc
# it gives int values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df.head(30)
```

```python
# used when categories have no order like A,B,C specific order
# Label Encoder gives random values like 1,2,3 for small,medium,high etc
# it gives float value
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder()
df[['Embarked']] = oe.fit_transform(df[['Embarked']])
```

```python
# used when label + binary number conversion
# it gives binary value
# binary encoder is not in sklearn, for this we need to install category_encoders
!pip install category_encoders
import category_encoders as ce

be = ce.BinaryEncoder(cols=['Embarked'])
df = be.fit_transform(df)
```
