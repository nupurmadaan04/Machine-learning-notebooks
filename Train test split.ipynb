Train Test Split

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("titanic.csv")
```

```python
# TRAIN TEST SPLIT - process in machine learning of dividing data into a training set to build the model and a test set to evaluate it on unseen data.
from sklearn.model_selection import train_test_split

# X = features, y = target
X = df[['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch']]
y = df['Survived']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,   # 20% test data
    random_state=42  # for reproducibility
)
```
