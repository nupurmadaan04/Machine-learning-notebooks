```python
import numpy as np
```

```python
arr = np.array([1,2,4,8,2,8,1,1,9,6,5])
```

```python
arr_mean = np.mean(arr)
```

```python
print(arr_mean)
```

```python
arr_median = np.median(arr)
```

```python
print(arr_median)
print(int(arr_median))
```

```python
from scipy import stats
arr_mode = stats.mode(arr)
print(arr_mode)
print(arr_mode.mode)
```

```python
mad = np.mean(np.abs(arr-arr_mean))
print(mad)
```
